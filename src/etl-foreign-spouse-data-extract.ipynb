### **Dataset3: 外籍配偶人數 (Local CSV)**

#### **Data Load**

```python
# Load 2024 divorce population local CSV
import pandas as pd
df_foreign_spouse = pd.read_csv(r"C:\Users\user\Downloads\foreign_spouse.csv")
# print first 5 rows
print(df_foreign_spouse.head())

# print column names
print(df_foreign_spouse.columns)

# print data types
print(df_foreign_spouse.dtypes)
```

#### **Load into PostgreSQL**

```python
# Create connection from Python to PostgreSQL
from sqlalchemy import create_engine
import os

user = os.getenv("PG_USER")
password = os.getenv("PG_PASSWORD")
host = "localhost"
port = "5432"
database = "twmarriage"
engine = create_engine(
    f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}"
)
```

```python
# Load dataframe to PostgreSQL
df_foreign_spouse.to_sql(
    "raw_foreign_spouse",
    engine,
    if_exists = "replace",
    index = False
)
```
